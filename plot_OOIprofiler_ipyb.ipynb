```
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
!pip install cmocean
import cmocean as cm
```

```
url='https://raw.githubusercontent.com/leviner/InterdisciplinaryEvaluation/master/data/merged/Profiler.csv'
df = pd.read_csv(url, index_col=[0], parse_dates=[0])
```

```
df.columns
```

```
fig = plt.figure(figsize=(15,18))
#fig.set_dpi(600) #When you want to make nicer versions for posters, play around with this and the fig_size property above
axis_limits = [pd.to_datetime('2018-10-05'),pd.to_datetime('2018-10-20')]



ax2 = plt.subplot(711)
pt = ax2.scatter(df.index,df.ctdpf_ckl_seawater_pressure,c=df.ctdpf_ckl_seawater_temperature,cmap=cm.cm.thermal,vmin=10,vmax=20)
ax2.invert_yaxis()
ax2.set_ylabel('Depth (m)')
ax2.set_xlim(axis_limits)
cbar = plt.colorbar(pt, ax=ax2)
cbar.set_label('T ($^{\circ}$C)')
ax2.grid()
ax2.xaxis.set_ticklabels([])

ax2 = plt.subplot(712)
pt =ax2.scatter(df.index,df.ctdpf_ckl_seawater_pressure,c=df.practical_salinity,cmap=cm.cm.haline)
ax2.invert_yaxis()
ax2.set_ylabel('Depth (m)')
ax2.set_xlim(axis_limits)
cbar = plt.colorbar(pt, ax=ax2)
cbar.set_label('Salinity (psu)')
ax2.grid()
ax2.xaxis.set_ticklabels([])

ax2 = plt.subplot(713)
pt = ax2.scatter(df.index,df.ctdpf_ckl_seawater_pressure,c=df.dofst_k_oxygen_l2,cmap=cm.cm.ice,vmin=150,vmax=250)
ax2.invert_yaxis()
ax2.set_ylabel('Depth (m)')
ax2.set_xlim(axis_limits)
cbar = plt.colorbar(pt, ax=ax2)
cbar.set_label('DO ($\mu$mol kg$^{-1}$)')
ax2.grid()
ax2.xaxis.set_ticklabels([])

ax2 = plt.subplot(714)
pt = ax2.scatter(df.index,df.ctdpf_ckl_seawater_pressure,c=df.fluorometric_cdom,cmap=cm.cm.turbid)
ax2.invert_yaxis()
ax2.set_ylabel('Depth (m)')
ax2.set_xlim(axis_limits)
cbar = plt.colorbar(pt, ax=ax2)
cbar.set_label('CDOM (ppb)')
ax2.grid()
ax2.xaxis.set_ticklabels([])

ax2 = plt.subplot(715)
pt = ax2.scatter(df.index,df.ctdpf_ckl_seawater_pressure,c=df.fluorometric_chlorophyll_a,cmap=cm.cm.algae,vmin=.01, vmax=12)
ax2.invert_yaxis()
ax2.set_ylabel('Depth (m)')
ax2.set_xlim(axis_limits)
cbar = plt.colorbar(pt, ax=ax2)
cbar.set_label('Chl-a ($\mu$g L$^{-1}$)')
ax2.grid()
ax2.xaxis.set_major_formatter(mdates.DateFormatter('%d %b %Y'))
fig.autofmt_xdate()
fig.align_labels()
```
